```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

wine_df = train.append(test)

len(wine_df)
```

```python
wine_df.head()
```

```python
# TOP 5 ACTIONABLE INSIGHTS FROM THE DATA train.csv and test.csv
```

```python
# 1

#Top 10 Wine Producing Countries
Top_Selling_Countries = pd.DataFrame(wine_df.country.value_counts()).iloc[0:10]
Top_Selling_Countries['Rank'] = list(range(1,11))
Top_Selling_Countries['Country_Name'] = Top_Selling_Countries.index.values
Top_Selling_Countries.set_index('Rank',inplace = True)
Top_Selling_Countries.columns = ['Number Of Wines Produced','Country Name']

Top_Selling_Countries
```

```python
figure, ax = plt.subplots(figsize=(10,5))
sns.set(style="ticks")
ax = sns.barplot(y="Country Name", x="Number Of Wines Produced", data=Top_Selling_Countries)
plt.title("Top 10 Wine Producing Countries",fontsize = 25)
plt.ylabel("Country Name",fontsize = 22)
plt.xlabel("Number Of Wines Produced",fontsize = 22)
```

```python
# 2

#Top 10 Wine Critics
critics_list = pd.DataFrame(wine_df.user_name.value_counts()).iloc[0:10]
critics_list['Rank'] = list(range(1,11))
critics_list['critic_Name'] = critics_list.index.values
critics_list.set_index('Rank',inplace = True)
critics_list.columns = ['Number of reviews given','Wine Critic']
critics_list
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set(style="ticks")
ax = sns.barplot(y="Wine Critic", x="Number of reviews given", data=critics_list)
plt.title("Top 10 Wine Critics",fontsize = 25)
plt.ylabel("critic_Name",fontsize = 20)
plt.xlabel("Number of reviews given",fontsize = 20)
```

```python
# 3

#Top 10 Wine Producing Wineries
winery_list = pd.DataFrame(wine_df.winery.value_counts()).iloc[0:10]
winery_list['Rank'] = list(range(1,11))
winery_list['Winery_Name'] = winery_list.index.values
winery_list.set_index('Rank',inplace = True)
winery_list.columns = ['Number Of Wines Produced','Winery']
winery_list
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set(style="ticks")
sns.barplot(y="Winery", x="Number Of Wines Produced", data=winery_list)
plt.title("Top 10 Wineries Producing Wine",fontsize = 25)
plt.ylabel("Winery",fontsize = 20)
plt.xlabel("Number Of Wines Produced",fontsize = 20)
```

```python
# 4

#Top 10 Wine Varieties based on Points given
top_variety = pd.DataFrame(wine_df['variety'].value_counts()[0:10])
top_variety['Rank'] = list(range(1,11))
top_variety['Variety_Name'] = top_variety.index.values
top_variety.set_index('Rank',inplace = True)
top_variety.columns = ['Total Points','Variety_Name']
top_variety
```

```python
varieties_points = wine_df[['points','variety']]
varieties_points = varieties_points[varieties_points.variety.isin(top_variety.Variety_Name)]
fig, ax = plt.subplots(figsize=(22,15))
sns.set_style("ticks")
sns.countplot(x="variety", data=varieties_points)

plt.ylabel("Points",fontsize = 25)
plt.xlabel("Varieties",fontsize = 25)
plt.title("Top 10 Wine Varieties based on Points given",fontsize = 30)
```

```python
# 5

# Distribution of Points For Top 10 Wine Varieties
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Pinot Noir'].groupby('points').count()).iloc[:,0]
variety_distribution = r
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Chardonnay'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Cabernet Sauvignon'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Red Blend'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Bordeaux-style Red Blend'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Riesling'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Sauvignon Blanc'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Syrah'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Rosé'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)
r = pd.DataFrame(wine_df[wine_df['variety'] == 'Merlot'].groupby('points').count()).iloc[:,0]
variety_distribution = pd.concat([variety_distribution,r],axis = 1)

variety_distribution.columns = ['Pinot Noir','Chardonnay','Cabernet Sauvignon','Red Blend','Bordeaux-style Red Blend',
                                'Riesling','Sauvignon Blanc','Syrah','Rosé','Merlot']
variety_distribution
```

```python
variety_distribution.plot.bar(stacked = True,figsize = (15,10),grid = False)
plt.title("Distribution of Points For Top 10 Wine Varieties",fontsize = 25)
plt.xlabel("Points",fontsize = 20)
```

```python
# 6

# Distribution Of Wine Prices
def price_div(price):
    if(price>0 and price<50):
        return 1
    elif(price >= 50 and price < 100):
        return 2
    elif(price >= 100 and price < 150):
        return 3
    elif(price >= 150 and price < 500):
        return 4
    else:
        return 5
```

```python
wine_df['Price_Group'] = wine_df['price'].apply(price_div)
count = list(wine_df['Price_Group'].value_counts())

label = ['Price <50','Price Between 50 & 100','Price Between 100 & 150','Price Between 150 & 500','Price > 500']
color = ['brown', 'green', 'coral', 'blue','red']
plt.figure(figsize = (15,8)) 
plt.pie(count,labels=label, colors=color,autopct='%1.2f%%', shadow=False, startangle=None)
plt.title("Distribution Of Wine Prices",fontsize = 25)
plt.axis('equal')
plt.show()
```
